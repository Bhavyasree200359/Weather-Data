```python
df_melted = pd.melt(df, id_vars=['YEAR'], var_name='Month', value_name='Temperature')
df_melted['Date'] = pd.to_datetime(df_melted['YEAR'].astype(str) + '-' + df_melted['Month'] + '-01', format='%Y-%b-%d')
df_melted = df_melted.rename(columns={'YEAR': 'Year'})
df_melted.head()
```

```python
print(df_melted.tail())
```

```python
months=df_melted['Month'].unique()
print(months)
data = [df_melted['Temperature'][df_melted['Month'] == month].dropna() for month in months]
plt.boxplot(data,labels=months)
plt.show()
print(data)
```

```python
outlier_counts = {}

for month in months:
    data = df_melted['Temperature'][df_melted['Month'] == month].dropna()
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    outliers = data[(data > upper_bound) | (data < lower_bound)]
    outlier_counts[month] = len(outliers)
for month, count in outlier_counts.items():
    print(f"Month: {month}, Outlier Count: {count}")
```

```python
monthly_means=df_melted.groupby("Month")["Temperature"].mean().reset_index()
plt.bar(monthly_means["Month"],monthly_means["Temperature"],color="#1e7b69")
```

```python
monthly_means = []

for month in df.columns[1:]: # Start from the third column to skip "YEAR"
    mean_value = df[month].mean()
    monthly_means.append([month, mean_value])

df_monthly_means = pd.DataFrame(monthly_means, columns=["Month", "Mean"])
df_monthly_means
x_values=df_monthly_means["Month"]
y_values=df_monthly_means["Mean"]
plt.barh(x_values,y_values,color="#123456")
plt.show()
```

```python
# Convert the monthly_means list into a DataFrame
df_monthly_means = pd.DataFrame(monthly_means, columns=["Month", "Temperature"])

# Now you can access columns using their names
plt.pie(df_monthly_means['Temperature'], labels=df_monthly_means["Month"], autopct="%1.1f%%")
plt.show()
```
